```python
from pydriller import RepositoryMining
import sys
import csv
import datetime 

#Commit hash, Author name, dATE,day, changeType, complexity, added, diff
filename="Npp_2.csv"
fields=['hash','name','date','day','changeType','added','DIF']
row=[]
with open(filename, 'w',newline='') as csvfile:
    csvwriter=csv.DictWriter(csvfile,fieldnames=fields)
    csvwriter.writeheader()
    
for commit in RepositoryMining('https://github.com/notepad-plus-plus/notepad-plus-plus.git').traverse_commits(): 
    for m in commit.modifications:
        hash_t=commit.hash
        name_t=commit.author.name
        date_t=commit.author_date
        ctype=m.change_type.name
        tadd=m.added
        dif=m.diff
        #print(type(Tdate))
        temp=str(date_t)
        #to calculate result on basis of weekdays not weekend i.e not on sat and sun.
        temp1=temp[:-3]
        temp2=temp[-2:]
        temp3=temp1+temp2
        date_obj=datetime.datetime.strptime(temp3, '%Y-%m-%d %H:%M:%S%z')
        day_t= date_obj.strftime('%A')
        rows=[{'hash': hash_t,'name': name_t,'date': date_t,'day': day_t,'changeType': ctype,'added':tadd,'DIF':dif}]
        with open(filename, 'a',encoding="utf-8",newline='') as csvfile:
            #csvwriter=csv.writer(csvfile)
            csvwriter=csv.DictWriter(csvfile,fieldnames=fields)
            #writing the data rows
            csvwriter.writerows(rows)
pd.read_csv('Npp_2.csv')
```
